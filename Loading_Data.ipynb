# Loading excel file where data is located on the 2nd sheet named **OECD.Stat export** starts from the **4th row** & the required columns are located in **A, C to T** columns of the excel sheet!!

```python
df=pd.read_excel('/content/GDPpercapita.xlsx',sheet_name='OECD.Stat export',skiprows=4,skipfooter=1,usecols="A,C:T")
df
```

```python
df.rename(columns={'Year':'metro'},inplace=True)
df
```

#Find if data have extra space in the beginning and end

```python
def beg_end_space(x):
  beg_space=x.str.startswith(' ').any()
  end_space=x.str.endswith(' ').any()
  print(f'have space in beginning: {beg_space} \nhave space in end: {end_space}')
```

```python
for col in df.columns:
  print(f"{col} : {beg_end_space(df[col])} '\n'")
```

#Fix extra space problem

```python
def strip_extra_space(x):
  striped_data=x.str.strip()
  return striped_data
```

```python
df['metro']= strip_extra_space(df.metro)
df
```

```python
beg_end_space(df.metro)
```

#Convert all columns excluding metro which is the 1st column, to numeric

```python
for col in df.columns[1:]:
  df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
df.dtypes
```

```python
for col in df.columns[1:]:
  df.rename(columns={col:'pcGDP'+col},inplace=True)
```

```python
df.describe(percentiles=[.25,.5,.99],include='all')
```

#Drop null based on subset

```python
df.dropna(subset=df.columns[1:],how='all',inplace=True)
```

```python
creator=json_df[:1].creators[0]
pprint.pprint(creator)
```
